```python
# Jumbled Game
import random
```

```python

lst= ['python', 'machinelearning', 'artificialintelligance', 'react', 'dotnet']
word= random.choice(lst)

jumble= ''.join(random.sample(word, len(word)))

print(jumble)

guess= input(f'Apna word daalo: ')

if(guess == word):
  print(f'Tum jeet gaye:')
else:
  print(f'Haar gaye kal aana:')
```
